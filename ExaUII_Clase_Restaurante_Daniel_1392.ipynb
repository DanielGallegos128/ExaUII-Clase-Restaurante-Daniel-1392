### Crear la clase "Restaurante"
### Con 3 atributos (nombre, especialidad, calificacion)
### 2 funciones leerDatos(), mostrarDatos()
### 2 objetos (local1, local2).

![texto del vínculo](https://raw.githubusercontent.com/DanielGallegos128/ExaUII-Clase-Restaurante-Daniel-1392/refs/heads/main/InfoClassRestaurante.png)

```python
class Restaurante:
    def __init__(self, nombre, especialidad, calificacion):
        self.nombre = nombre
        self.especialidad = especialidad
        self.calificacion = calificacion

    def leerDatos(self):
        self.nombre = input("Inserte el nombre del restaurante: ")
        self.especialidad = input("Inserte la especialidad del restaurante: ")
        self.calificacion = input("Inserte la calificacion del restaurante: ")

    def mostrarDatos(self):
        print(f"Nombre: {self.nombre}")
        print(f"Especialidad: {self.especialidad}")
        print(f"Calificacion: {self.calificacion}")
        return self.nombre, self.especialidad, self.calificacion

# --- Creación y uso de los 2 objetos ---

print("Local 1")
local1 = Restaurante("", "", "") # Creando objeto local 1
local1.leerDatos()
local1.mostrarDatos()
print("__________________")

print("Local 2")
local2 = Restaurante("", "", "") # Creando objeto local 2
local2.leerDatos()
local2.mostrarDatos()
print("__________________")

print("Programa realizado por Daniel Gallegos Nl=20")
```
